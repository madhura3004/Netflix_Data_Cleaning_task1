```python
#Import Libraries & Load the Dataset
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/archive.zip")

# Basic info
print("Shape:", df.shape)
df.head()
```

```python
# Count missing values in each column
print(df.isnull().sum())

# Option 1: Drop rows with missing values
df_cleaned = df.dropna()


```

```python
# Check for duplicate rows
print("Duplicates:", df_cleaned.duplicated().sum())

# Drop duplicates if any
df_cleaned = df_cleaned.drop_duplicates()
```

```python
# Convert 'type' and 'country' to lowercase and strip whitespace
df_cleaned['type'] = df_cleaned['type'].str.lower().str.strip()
df_cleaned['country'] = df_cleaned['country'].str.lower().str.strip()
```

```python
# Convert 'date_added' to datetime format
df_cleaned['date_added'] = pd.to_datetime(df_cleaned['date_added'], errors='coerce')
```

```python
# Rename all columns to lowercase and replace spaces with underscores
df_cleaned.columns = df_cleaned.columns.str.lower().str.replace(' ', '_')
```

```python
# Check column data types
print(df_cleaned.dtypes)

# Fix types if needed (example: release_year should be int, date_added is datetime)
df_cleaned['release_year'] = df_cleaned['release_year'].astype(int)
```

```python
df_cleaned.to_csv("cleaned_netflix_dataset.csv", index=False)
```
